This function is used to track the number of indicies(rows) a particular car takes in our data
In other words, it is the number of times a particular car appears in different frames

```python
def index_of_first_distinct_number(datt,vfirst,last):
    indx=0
    datt = datt
    sx = len(datt)
    for jj in range(last,sx):      
        if (datt[jj,1] == vfirst[0,1]):
            indx = jj 
        else:
            break

    return indx
        #assume an immaginary number beyond the last value 
```

This function is used to create a matrix for each cars with their area, entry point and exit point

```python
def function_distinct_blocks_of_data(datt):
    sx = len(datt)
    
    skip = 0
    mdd = []
    last=0
    
    jk  = 0
    ji = 0
    while ji < sx:
        if len(datt[ji]) != 0:
            jk = jk+1
            ifirst = ji;
            vfirst = datt[ji]
            
            indx = index_of_first_distinct_number(datt,vfirst,last)
            ilast = indx
            last =ilast+1
            
            mdd.append([vfirst[0,1], ifirst, ilast])
            
            skip = 0
            ji =ilast+1
            
            
        else:
            skip = skip+1       
            ji = ji+1
        
    return np.matrix(mdd)
            
```
