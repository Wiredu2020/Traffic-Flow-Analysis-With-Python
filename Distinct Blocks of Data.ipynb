```python
def function_distinct_blocks_of_data(datt):
    sx = len(datt)
    jk  = 0
    ji = 0
    skip = 0
    mdd = []
    last=0
    while ji < sx:
        if len(datt[ji]) != 0:
            jk = jk+1
            ifirst = ji;
            vfirst = datt[ji]
            
            indx = index_of_first_distinct_number(datt,vfirst,last)
            ilast = indx
            last =ilast+1
            
            mdd.append([vfirst[0,1], ifirst, ilast])
            
            skip = 0
            ji =ilast+1
            
            
        else:
            skip = skip+1       
            ji = ji+1
        
    return np.matrix(mdd)
            
```
