```python
'''
PURPOSE: 
   After getting the area and centroid of each car 
   at each time, this code is used to arrange all 
   cars of a particular area from entry to exit in first 
   few (say 5) rows, the car that follows to be in the next 
   few (say 6-10) rows, etc.
 INPUT
 data -> data obtained from the script:
    TrackCars_manual_differentAreas_Analysis.m
 OUTPUT
 datt -> with vehicles arranged in order
 AUTHOR: Dr. Joseph K. Ansong
 DATE: 3 Dec. 2019.

'''

def function_arrange_cars(data, halfwidth):
    hwidth = halfwidth;
    dataMM = data
    datt = []
    #to be converted
    
    if dataMM[0,0] != nan:
        jx =1
    else:
        jx =0
    if jx == 0:
        dataMM[0,:]= []
    
    (sx, ny) = dataMM.shape
    
    
    fv=dataMM[0,1]
    
    datt.append(dataMM[0,0:ny])
    kj = 0
    ii=1
    idxx =[0]
    while ii < sx:
        
        if (dataMM[ii,1]==fv) and  (dataMM[ii,2]>= datt[:,2]):
            kj+=1
            datt.append(dataMM[ii,:])
            idxx.append(kj)
            ii+=1
        
         #In case of a jam, it is possible to mistakenly click a location
         #which shows that the vehicle moved backwards. The ff is to use the
         #same position as the previous location. The last condition ensures 
         #that a car with the same dimensions will not be used since it also 
         #satisfies the second condition.
        
        elif (dataMM[ii,1]== fv) and (dataMM[ii,2] < datt[:,3]) and (abs(dataMM[ii,2]-datt[:,2]) <= hwidth ):
            
            
            kj+=1
            datt.append(dataMM[ii-1,:])
            idxx.append(kj)
            ii+=1
            
        elif (dataMM[ii,1]==fv) and (dataMM[ii,2] <datt[:,2]) and (abs(dataMM[ii,2]-datt[:,3])> 5*hwidth):
           
                
            dataMM[idxx[kj],:] = []
              #====index of first Non-NaN value

            if dataMM[0,0] != nan:
                jx =1
            else:
                jx =0
            if jx == 0:
                dataMM[0,:]= []

            (sx, ny) = dataMM.shape


            fv=dataMM[0,1]

            datt.append(dataMM[0,0:ny])
            kj = 0
            ii=1
            idxx =[0]
        
            #====index of first Non-NaN value
        
        elif ii==sx:
    
            dataMM[idxx[kj],:] =[];
        

            #====index of first Non-NaN value

            if dataMM[0,0] != nan:
                jx =1
            else:
                jx =0
            if jx == 0:
                dataMM[0,:]= []

            (sx, ny) = dataMM.shape


            fv=dataMM[0,1]

            datt.append(dataMM[0,0:ny])
            kj = 0
            ii=1
            idxx =[0]

        elif (len(dataMM)==0):
        #size(dataMM)
             break
        else:
            ii=ii+1;
   
    return datt
            
            
```
